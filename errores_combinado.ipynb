## 1. Sintaxis básica: `try-except`

Es el mecanismo principal para evitar que un programa se detenga abruptamente.

- `try`: aquí va el código que puede fallar.
- `except`: aquí decides qué hacer si ocurre un error.

### Errores comunes
- `ValueError`: el tipo parece correcto, pero el contenido no lo es.
- `ZeroDivisionError`: dividir por cero.
- `TypeError`: mezclar tipos incompatibles.
- `IndexError`: acceder a una posición que no existe.


```python
# Ejemplo 2: capturar errores frecuentes
try:
    number1 = int(input("Enter a number as a base: "))
    number2 = int(input("Enter another number as a divisor: "))
    result = number1 / number2
    print(f"The result of {number1} divided by {number2} is {result}")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Ejemplo 3: otros errores típicos
try:
    text = "Hello, World! "
    number = 5
    combined = text + number
    print(combined)
except TypeError:
    print("Error: Cannot concatenate string and integer.")

try:
    my_list = [1, 2, 3]
    print(my_list[5])
except IndexError:
    print("Error: Index is out of range.")
```

## 2. El ciclo completo: `else` y `finally`

Podemos extender la estructura para tener más control:

- `else`: se ejecuta solo si **no** hubo error.
- `finally`: se ejecuta siempre, haya error o no.

`finally` es útil para limpieza; `else` ayuda a separar el caso exitoso del caso con error.


```python
# Ejemplo 4: else y finally
try:
    result = 10 / 5
except ZeroDivisionError:
    print("Cannot divide by zero.")
else:
    print(f"Operation successful. Result: {result}")
finally:
    print("Block finished.")

print("Continuing with the rest of the program.")
```

## 3. Validación con `raise`

Usamos `raise` cuando el código es válido para Python, pero inválido para nuestra lógica.

Ejemplo típico: edad negativa, precio negativo o mes fuera de rango.


```python
def set_age(age: int):
    if age <= 0:
        raise ValueError("Age must be a positive integer.")
    print(f"Age set to {age}")

try:
    set_age(int(input("Enter your age: ")))
except ValueError as e:
    print(f"Validation Error: {e}")
```

## 4. Patrón muy útil: input robusto con `while True`

Cuando pides datos al usuario, no quieres que el programa muera a la primera entrada incorrecta.

Este patrón es muy común en scripts y pequeños menús por terminal.


```python
def get_valid_int():
    while True:
        try:
            return int(input("Enter a number: "))
        except ValueError:
            print("Invalid input. Try again.")

number = get_valid_int()
print(f"You entered: {number}")
```
