### Ejemplo 4.1 (rcParams + rotación)

```python
plt.rcParams.update({
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10
})

day_mean = tips.groupby('day', as_index=False)['total_bill'].mean().sort_values('total_bill')
#Agrupa por dia en el groupby day

fig, ax = plt.subplots(figsize=(7,3.5))
ax.barh(day_mean['day'], day_mean['total_bill'])  #Filtro por total bill y saco las medias, queremos la media por dia
ax.set_title('Average total_bill by day')
ax.set_xlabel('average total_bill')
plt.tight_layout()
plt.show()
```

## 5) Comparación categórica (Seaborn): barplot, order, hue, error bars

Seaborn `barplot`:
- Por defecto dibuja la **media** de `y` por cada categoría en `x`.
- Puede añadir intervalos (según versión) automáticamente.

Cosas que importan en la práctica:
- Ordenar categorías (`order=`)
- Comparar subgrupos con `hue=`
- Elegir paleta (`palette=`)


### Ejemplo 5.1 (barplot con orden)
Ordenamos por media de total_bill y pintamos.

```python
order_days = tips.groupby('day')['total_bill'].mean().sort_values().index

plt.figure(figsize=(7,4))
sns.barplot(data=tips, x='day', y='total_bill', order=order_days, palette='Set2') #order es para ordenar el grafico, palette es para cambiar el color
plt.title('Average total_bill by day (ordered)')
plt.tight_layout()
plt.show()
```

### Ejemplo 5.2 (barplot con hue)
Comparación por turno (`time`).

```python
plt.figure(figsize=(7,4))
sns.barplot(data=tips, x='day', y='tip', hue='time', palette='muted') #hue es para diferenciar los colores, cambia el color segun el time
plt.title('Average tip by day split by time')
plt.tight_layout()
plt.show()
```

## 9) Series temporales: líneas, media móvil, estacionalidad (flights)

Aquí el objetivo es contar una historia:
- tendencia a largo plazo
- estacionalidad (patrón que se repite)
- variabilidad

Primero convertimos `year+month` a datetime mensual.

### Ejemplo 9.1 (serie mensual + rolling 12 meses)

```python
fl = flights.copy()

# Convert month name to month number safely
fl["month_num"] = pd.to_datetime(fl["month"], format="%b").dt.month

# Build datetime column
fl["date"] = pd.to_datetime(
    dict(year=fl["year"], month=fl["month_num"], day=1)
)

# Sort
fl = fl.sort_values("date")

# Rolling mean
fl["rolling_12"] = fl["passengers"].rolling(12, min_periods=1).mean()

# Plot
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(fl["date"], fl["passengers"], alpha=0.5, label="monthly") #estos son los datos originales por mes
ax.plot(fl["date"], fl["rolling_12"], linewidth=2, label="rolling 12m") #Esta es la media movil por cada año

ax.set_title("Flights passengers: monthly + rolling 12m")
ax.set_xlabel("date")
ax.set_ylabel("passengers")
ax.legend()

plt.xticks(rotation=30) # Rotar las etiquetas de los años
plt.tight_layout() #Ajusta el espaciado y la posicion para mejorar la presentacion
plt.show()
```

### Ejercicio 7

1) Crea `pct_change` mes a mes y dibuja un gráfico de líneas de ese crecimiento.
2) Dibuja la serie anual (agrega por año) y compárala con la mensual.
3) Comenta 2 cosas: tendencia y estacionalidad.

```python
# Ejercicio 7
# Copia del DataFrame original para realizar cambios y visualizaciones
fl_ex = fl.copy()

# Cambio porcentual mensual (MoM percent change)
# Calcula el cambio porcentual de los pasajeros entre cada mes
fl_ex['pct_change'] = fl_ex['passengers'].pct_change() * 100

# Crea un gráfico de línea para mostrar el cambio porcentual mensual
# Define el tamaño del gráfico
fig, ax = plt.subplots(figsize=(10, 4))

# Dibuja la línea del cambio porcentual mensual
ax.plot(fl_ex["date"], fl_ex["pct_change"], alpha=0.5, label="monthly")

# Configura el título del gráfico
ax.set_title("Cambio en porcentaje")

# Configura las etiquetas del eje x y el eje y
ax.set_xlabel("date")
ax.set_ylabel("pct_change")

# Agrega una leyenda al gráfico
ax.legend()

# Rotar las etiquetas del eje x en un ángulo de 30 grados
plt.xticks(rotation=30)

# Ajusta automáticamente el espaciado y la posición de los elementos
plt.tight_layout()

# Muestra el gráfico
plt.show()

# Agrupa los datos del DataFrame por año, calcula la suma de los pasajeros
fl_anual = fl_ex.groupby("year", as_index=False).agg({"passengers": "sum"})

# Convierte la columna "year" en formato de fecha
fl_anual["date"] = pd.to_datetime(dict(year=fl_anual['year'], month=1, day=1))

# Crea un gráfico de línea para mostrar la comparación anual vs mensual
# Define el tamaño del gráfico
fig, ax = plt.subplots(figsize=(10, 4))

# Dibuja la línea de los pasajeros mensuales
ax.plot(fl_ex["date"], fl_ex["passengers"], alpha=0.5, label="monthly")

# Dibuja la línea de los pasajeros anuales
ax.plot(fl_anual["date"], fl_anual["passengers"], alpha=0.5, label="yearly", marker="o")

# Configura el título del gráfico
ax.set_title("Comparación pasajeros anuales vs mensuales")

# Configura las etiquetas del eje x y el eje y
ax.set_xlabel("date")
ax.set_ylabel("passengers")

# Agrega una leyenda al gráfico
ax.legend()

# Rotar las etiquetas del eje x en un ángulo de 30 grados
plt.xticks(rotation=30)

# Ajusta automáticamente el espaciado y la posición de los elementos
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

## 10) Heatmaps: matriz mes×año (flights)

Heatmap útil para ver estacionalidad:
- filas: mes
- columnas: año
- color: pasajeros

Importante: un heatmap no es para "precisión"; es para patrones.

### Ejemplo 10.1 (heatmap base)

```python
pivot = fl.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')


plt.figure(figsize=(10,5))
sns.heatmap(pivot, cmap='viridis')
plt.title('Heatmap: passengers (month × year)')
plt.tight_layout()
plt.show()
```

### Ejemplo 10.2 (heatmap con anotaciones)
Ojo: `annot=True` puede saturar si la matriz es grande. Aquí es razonable.

```python
plt.figure(figsize=(10,5))
sns.heatmap(pivot, cmap='magma', annot=True, fmt='.0f') #fmt sirve para el formato, en este caso  especifica que los valores deben formatearse como números enteros 
                                                        #El . indica que se debe utilizar el formato de punto decimal, y 0f indica que se deben redondear los números a cero decimales.
plt.title('Heatmap: passengers (annotated)')
plt.tight_layout()
plt.show()
```

## 11) Anotaciones, etiquetas y storytelling

Un gráfico útil suele tener:
- título que diga la conclusión o el tema
- ejes con unidades
- leyenda si aporta
- anotaciones SOLO cuando clarifican

Vamos a anotar el punto máximo en la serie de flights.

### Ejemplo 11.1 (anotar máximo)
Anotación simple con `ax.annotate`.

```python
fl2 = fl.copy()
idx = fl2['passengers'].idxmax()
max_row = fl2.loc[idx]

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(fl2['date'], fl2['passengers'], alpha=0.7)
ax.set_title('Flights passengers (max annotated)')
ax.set_xlabel('date')
ax.set_ylabel('passengers')

ax.scatter([max_row['date']], [max_row['passengers']], s=60)
ax.annotate(
    f"Max: {int(max_row['passengers'])}",
    xy=(max_row['date'], max_row['passengers']),
    xytext=(max_row['date'], max_row['passengers']*0.9),
    arrowprops={'arrowstyle': '->'}
)

plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
