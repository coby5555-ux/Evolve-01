## 1️⃣ Series: la estructura más simple

Una **Series** es un array unidimensional con índice.

Puedes verla como una columna suelta con etiquetas.


```python
values = [10, 20, 30, 40]
s_list = pd.Series(values, index=['a', 'b', 'c', 'd'])

data_dict = {'Apple': 100, 'Banana': 200, 'Cherry': 150}
s_dict = pd.Series(data_dict)

print("Series from list:", s_list)
print()
print("Series from dict:", s_dict)
print()
print("Index:", s_dict.index)
print("Values:", s_dict.values)
```

## 2️⃣ DataFrames: la tabla de datos

Un **DataFrame** es una estructura bidimensional con filas y columnas.

Normalmente se crea a partir de un diccionario de listas.


```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Paris', 'London', 'Tokyo'],
    'Salary': [50000, 60000, 70000, 80000],
}

df_manual = pd.DataFrame(data)
display(df_manual)
print("Shape:", df_manual.shape)
```

## 6️⃣ Filtrado y transformación

El patrón más común en trabajo real es:
1. filtrar filas,
2. crear columnas nuevas,
3. volver a filtrar,
4. quedarte solo con lo útil.


```python
mask = df_large['A'] > 50
filtered_df = df_large[mask]

print("Original rows:", df_large.shape[0])
print("Filtered rows:", filtered_df.shape[0])

complex_filter = df_large[(df_large['A'] > 50) & (df_large['B'] < 20)]
display(complex_filter.head())

df_large['Suma'] = df_large['A'] + df_large['B']
display(df_large.head())
```

## 7️⃣ Ordenación

Después de seleccionar y filtrar, normalmente queremos ordenar resultados.

Podemos ordenar por una o varias columnas, y decidir el sentido de cada una.


```python
df_sorted = df_large.sort_values(by='A', ascending=True)
df_multi_sort = df_large.sort_values(by=['A', 'B'], ascending=[False, True])

display(df_sorted.head())
display(df_multi_sort.head())
```
