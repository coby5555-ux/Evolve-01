## 1. Introducción a POO

La programación orientada a objetos organiza el código alrededor de **objetos**.

### Conceptos clave
- **Clase**: la plantilla.
- **Objeto**: una instancia concreta de esa plantilla.
- **Atributo**: la información del objeto.
- **Método**: lo que el objeto sabe hacer.

Piensa así:
- la clase describe,
- el objeto existe,
- los atributos guardan estado,
- los métodos expresan comportamiento.


```python
# Example 1: first class and first object
class Perro:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def ladrar(self, sonido="Guau"):
        print(f"{self.nombre} dice: {sonido}")


mi_perro = Perro("Fido", 3)
print(f"El perro se llama {mi_perro.nombre} y tiene {mi_perro.edad} años.")
mi_perro.ladrar()
```

## 2. Atributos, métodos y `__init__`

El método `__init__` se ejecuta automáticamente cuando creamos un objeto.

Sirve para inicializar sus atributos.

Los métodos usan `self` para acceder a los datos del propio objeto.


```python
import math


class Circulo:
    def __init__(self, radio):
        self.radio = float(radio)
        self.diametro = self.radio * 2

    def area(self):
        return math.pi * (self.radio ** 2)

    def perimetro(self):
        return 2 * math.pi * (self.radio)

mi_circulo = Circulo(6)
print("Radio:", mi_circulo.radio)
print("Área:", mi_circulo.area())
print("Diámetro:", mi_circulo.diametro)
print("Perímetro:", round(mi_circulo.perimetro(), 2))
```

### Ejercicio 2: clase `Rectangulo`

Crea una clase `Rectangulo` con:

- atributos: `base` y `altura`
- método: `area()`, que devuelva `base * altura`

Después crea una instancia y calcula el área.


```python
# Exercise 2

class Rectangulo:
    def __init__(self, base, altura):
        self.base = float(base)
        self.altura = float(altura)

    def area(self):
        return self.base * self.altura

mi_rectangulo = Rectangulo(15, 20)

print(f"El área es: {mi_rectangulo.area()}") 
```

## 3. Herencia

La herencia permite reutilizar código y crear jerarquías.

Una clase hija puede:
- heredar atributos y métodos,
- añadir atributos nuevos,
- sobrescribir métodos.

`super()` permite aprovechar el constructor o comportamiento de la clase padre.


```python
class Animal:
    def __init__(self, nombre, patas=4):
        self.nombre = nombre
        self.patas = patas

    def hablar(self):
        print(f"{self.nombre} hace un sonido.")


class Perro(Animal):
    def __init__(self, nombre, raza):
        super().__init__(nombre)
        self.raza = raza

    def hablar(self):
        print(f"{self.nombre} dice: ¡Guau!")


class Gato(Animal):
    def __init__(self, nombre, color):
        super().__init__(nombre)
        self.color = color

    def hablar(self):
        print(f"{self.nombre} dice: Miau")


perro = Perro("Rex", "Labrador")
gato = Gato("Michi", "Blanco")
perro.hablar()
gato.hablar()
```

### Ejercicio 3: `Vehiculo` y `Coche`

1. Crea una clase `Vehiculo` con:
   - atributos: `marca`, `modelo`
   - método: `descripcion()`

2. Crea una clase `Coche` que herede de `Vehiculo` y añada:
   - atributo: `puertas`
   - sobrescritura de `descripcion()` para incluir el número de puertas


```python
# Exercise 3

class vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def descripcion(self):
        return f"Marca: {self.marca}, Modelo: {self.modelo}"
    
class coche(vehiculo):
    def __init__(self, marca, modelo, puertas):
        super().__init__(marca, modelo)
        self.puertas = puertas

    def descripcion(self):
        print(f"vehiculo marca", self.marca, "modelo", self.modelo, "con", self.puertas, "puertas")

mi_coche = coche("ford", "focus", 5)
mi_coche.descripcion()
```

## 4. Métodos especiales: `__str__`

Los métodos especiales permiten integrar mejor nuestros objetos con Python.

Hoy nos quedamos con uno muy útil:

- `__str__`: define cómo se ve el objeto cuando usamos `print()`.

Hace que nuestros objetos sean más legibles y más fáciles de depurar.


```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"{self.nombre}, {self.edad} años"

    def descripcion(self):
        return f"Persona: {self.nombre}, Edad: {self.edad}"


persona = Persona("Lucía", 28)
print(persona)
print(persona.descripcion())
```

## 5. Ejercicio final: sistema de gestión de estudiantes

Ahora vamos a combinar todo.

Crea las siguientes clases:

### `Persona`
- atributos: `nombre`, `edad`
- método: `__str__`

### `Estudiante` (hereda de `Persona`)
- atributo extra: `carrera`
- atributo extra: `notas` (lista)
- método: `promedio()`
- `__str__` personalizado

### `Profesor` (hereda de `Persona`)
- atributo extra: `departamento`
- atributo extra: `asignaturas` (lista)
- método: `agregar_asignatura(nombre)`
- `__str__` personalizado

### Qué hacer después
- crea al menos dos estudiantes con notas
- crea un profesor
- añade asignaturas al profesor
- imprime toda la información


```python
# Final exercise

class Persona:

    def __init__(self, nombre, edad):
        self.nombre= nombre
        self.edad = int(edad)
    def __str__(self):
        return f"Me llamo {self.nombre} y tengo {self.edad} años"

class Estudiante(Persona):

        def __init__(self, nombre, edad, carrera, notas=[0]):
            super().__init__(nombre, edad)
            self.carrera = carrera
            self.notas = notas

        def promedio(self):
        #mejorar
            return sum(self.notas)/len(self.notas)
        
        def __str__(self):
            return f"Me llamo {self.nombre} soy estudiante y tengo {self.edad} años"

class Profesor(Persona):
    def __init__(self, nombre, edad, departamento, asignaturas=[]):
        super().__init__(nombre, edad)
        self.departamento = departamento
        self.asignaturas = asignaturas

    def agregar_asignatura(self, asignatura):
        self.asignaturas.append(asignatura)
        print(f"Se ha agregado la asignatura {asignatura}")

    def __str__(self):
        return f"Me llamo {self.nombre} soy profesor de las asignaturas {self.asignaturas}"

alumno_uno = Estudiante("david", 23, "Data science", [10, 10, 10])
alumno_dos = Estudiante("Jose", 20, "Ciber", [10, 9, 2.1])
profesor= Profesor("Nicolas", 30, "Python", ["Python", "SQL", "IA"])

print(profesor)

profesor.agregar_asignatura("JAVA")

print(alumno_uno)
print(alumno_dos)

profesor.agregar_asignatura("Algebra")

print(profesor)
print(alumno_uno.promedio())

alumno_dos.promedio()
```

## Reto opcional: la tienda

Si el grupo va rápido, puedes intentar este mini reto adicional.

Crea una clase `Tienda` con:
- `nombre`
- `direccion`
- `num_empleados`
- `ventas` (lista de 3 meses)

Y añade métodos para:
- calcular ventas totales
- obtener ventas del último mes
- calcular media mensual por empleado


```python
# Optional challenge

class Tienda:
    def __init__(self, nombre, direccion, empleados, ventas = []):
        self.nombre = nombre
        self.direccion = direccion
        self.empleados = empleados
        self.ventas = ventas
    
    def sales(self):
        return sum(self.ventas)

    def sales_last_month(self):
        return self.ventas[-1]

    def average_employee(self):
        lista = []
        for venta in self.ventas:
            lista.append(venta/self.empleados)
        return lista

mi_tienda = Tienda("Rubén", "micalle", 5, [50000, 15000, 123])

print(mi_tienda.sales())
print(mi_tienda.average_employee())
print(mi_tienda.sales_last_month())
```
